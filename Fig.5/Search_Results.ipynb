```python
## Non plotting libraries
import numpy as np 
import pandas as pd

# Plotting libraries
import matplotlib.pyplot as plt
import matplotlib
import scienceplots
import matplotlib.colors as mcolors
import matplotlib.cm as cm
from matplotlib.lines import Line2D
from matplotlib.colors import ListedColormap
plt.style.use('science')
matplotlib.rcParams.update({'font.size': 25})
```

Loading in search results

```python
low_f_results = np.loadtxt('../Shared_data_files/Search_results/low_f/final_best_positions.txt')
high_f_results = np.loadtxt('../Shared_data_files/Search_results/high_f/final_best_positions.txt')
search_results = []
search_results.extend(list(high_f_results[high_f_results[:,-1]>90]))
search_results.extend(list(low_f_results[low_f_results[:,-1]>90]))
search_results = np.array(search_results)
```

Loading in catalog

```python
catalog = pd.read_csv('../Shared_data_files/catalog.csv')
catalog['Name'] = np.array([['Source 1'],
                            ['Source 5'],
                            ['Source 8'],
                            ['Source 9'],
                            ['Source 10']])
catalog
```

```python
# Lists to store errors 
delta_mc = []
delta_lambda = []
delta_beta = []
delta_f_low = []
```

Work out fractional errors for each search result

# Full duty cycle results

```python

for index, sobbh in catalog.iterrows():
    print(sobbh['Name'])

    m1 = sobbh['Mass1']
    m2 = sobbh['Mass2']

    mc = ((m1*m2)**(3/5))/(m1+m2)**(1/5)
    f_low = sobbh['InitialFrequency']

    index = np.argmin(np.abs(search_results[:,0] - mc))

    max_upsilon = search_results[index,-1]
    
    print('Delta mc:',np.abs(search_results[index][0]-mc),np.abs(search_results[index][0]-mc)/mc)
    print('Delta f_low: ',np.abs(search_results[index][4]-f_low),np.abs(search_results[index][4]-f_low)/f_low)
    print('Max upsilon: ',max_upsilon)

    delta_mc.append(np.abs(search_results[index][0]-mc)/mc)
    delta_f_low.append(np.abs(search_results[index][4]-f_low)/f_low)
    
    
    print('\n')

        
```

# Gapped sources

```python
gapped_delta_mc = []
gapped_delta_f_low = []
for index,row in catalog.iterrows():
    print(row['Name'])
    final_swarm_data = np.loadtxt('../Shared_data_files/Gapped/' + row['Name'] + '/best_position_and_value.txt')
    best_position= final_swarm_data[:-1]
    best_value = final_swarm_data[-1]

    m1 = row['Mass1']
    m2 = row['Mass2']

    mc = ((m1*m2)**(3/5))/(m1+m2)**(1/5)

    f_low = row['InitialFrequency']

    print('Delta mc:',np.abs(best_position[0]-mc),np.abs(best_position[0]-mc)/mc)
    print('Delta f_low: ',np.abs(best_position[4]-f_low),np.abs(best_position[4]-f_low)/f_low)
    print('\n')

    gapped_delta_mc.append(np.abs(best_position[0]-mc)/mc)
    gapped_delta_f_low.append(np.abs(best_position[4]-f_low)/f_low)
```

```python
delta_mc = np.array(delta_mc).flatten()
delta_f_low = np.array(delta_f_low).flatten()
```

```python
# Choose a discrete colormap (here "tab10")
base_cmap = plt.cm.get_cmap("tab10")

# Extract the first 5 colors
colors = base_cmap.colors[:5]

# Create a ListedColormap
cmap = ListedColormap(colors)
```

```python
plt.figure(figsize=(8,5))

for i in range(delta_mc.size):

    plt.scatter(delta_f_low[i],delta_mc[i],color=cmap(i),marker='o',s=150)
    plt.scatter(gapped_delta_f_low[i],gapped_delta_mc[i],color=cmap(i),marker='v',s=150,facecolors='none')



plt.ylabel(r'$\frac{\Delta \mathcal{M}_c}{\mathcal{M}_c}$',fontsize=25)
plt.xlabel(r'$\frac{\Delta f_{\mathrm{in}}}{f_{\mathrm{in}}}$',fontsize=25)



labels = ['ID 1', 'ID 5', 'ID 8', 'ID 9', 'ID 10']
marker_style = 'D'        # diamond; swap for 'o', '^', '*', 's', etc.

handles = [Line2D([], [], marker='o', linestyle='',
                  color=col,markersize=12)   # edge so marker stands out
           for col in cmap.colors]

plt.legend(handles=handles,
          labels=labels,handletextpad=0.1,frameon=True,fontsize=20)

plt.semilogy()
plt.semilogx()
plt.grid()
plt.savefig('Fractional_errors.pdf')
plt.show()
```
